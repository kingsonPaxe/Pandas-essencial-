# Seleção de dados em um DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv('Info_Alunos.csv')
df
```

```python
df["Sexo"] = "M"
df
```

```python
# Adicionando uma mulher na tabela
df.loc[7] = ["Manuela", 25, "Comerciante", "manuela@gmail.com", "F"]
df
```

```python
dados = [
    {"Nomes": "Carla", "Idades": 23, "Profissão": "Cantora", "E-mail": "carla@gmail.com", "Sexo": "F"},
    {"Nomes": "Bruna", "Idades": 18, "Profissão": "Estudante", "E-mail": "bruna@gmail.com", "Sexo": "F"},
]

df = df._append(dados, ignore_index = True)
df
```

## Seleção de dados

* A aplicação de operadores lógicos em colunas juntamente com o indexador loc permite a seleção de dados de uma maneira bastante ágil.

* Como visto anteriormente, o resultado da aplicação de operadores lógicos em colunas é uma lista de booleanos representando as linhas que se adequam ao critério de seleção.

* O indexador loc permite utilizar como parâmetro uma lista com valores booleanos que representam as linhas que serão selecionadas.

* Exemplo de como selecionar do DataFrame as pessoas que já atingiram a maioridade penal.

```python
df[df.Sexo == "F"].shape[0] # Pesqueisando quantas mulheres do sexo femenino temos no dataframe
```

```python
# Agora eu quero saber quem são essas meninas
meninas = df[df.Sexo == "F"]
meninas
```

```python
# De todas elas quem é mais velha?
meninas.loc[meninas["Idades"] == meninas.Idades.max()]
```

```python
df[df["Idades"] == df[df["Sexo"] == "F"].Idades.max()] # Outra forma de saber quem é saber quem é mais velha
```

```python
df.loc[df.Idades == min(df.Idades)]
```

```python
# Exportar para csv
df = df.to_csv("info_alunos2.csv", index = False)
```
