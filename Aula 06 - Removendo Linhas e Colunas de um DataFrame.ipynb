# **1. Removendo Linhas e Colunas de um DataFrame**

```python
import pandas as pd
```

```python
dt = {'Nomes': {0: 'Cebolinha',
  1: 'Sandro',
  2: 'Fernando',
  3: 'Pedro',
  4: 'Konde',
  5: 'Cisco',
  6: 'Gerson'},
 'Idades': {0: 23, 1: 20, 2: 22, 3: 32, 4: 47, 5: 18, 6: 28},
 'Profissão': {0: 'Engenheiro Cívil',
  1: 'Estudante',
  2: 'Estudante',
  3: 'Estudante',
  4: 'Músico',
  5: 'Estudante',
  6: 'Estudante'},
 'E-mail': {0: 'cebolinha@gmail.com',
  1: 'sandro_santos@gmail.com',
  2: 'fernandoFM@gmail.com',
  3: 'pedropp@gmail.com',
  4: 'konde@gmail.com',
  5: 'francisco@gmail.com',
  6: 'Gerson@gmail.com'}}
df = pd.DataFrame(dt)
df
```

```python
df.loc[6,"Profissão"] = "Desenvolvedor Java"
df # Fiz uma alteração na profissão do Gerson
```

## drop()

* ###  É possível remover linhas ou colunas de um DataFrame utilizando o método drop.

* ### Alguns dos parâmetros do método drop são:
    * **index:** recebe um rótulo ou uma lista de rótulos das linhas que serão removidas.
    * **columns:** recebe um rótulo ou uma lista de rótulos das colunas que serão removidas.
    * **inplace:** determina se as mudanças devem ser aplicadas diretamente no DataFrame ou em uma cópia (valor padrão é False).


```python
df
```

## Apangando uma linha com o método drop()

```python
# Vamos apagar a linha do konde e do cebolinha
df.drop(index = [0,4], inplace = True)
df
```

## Apagando uma coluna usando o método drop()

```python
# vamos apagar a coluna de profissão
df.drop(columns = ["Profissão", "Idades"], inplace = True)
df
```

```python
# se eu quiser criar uma coluna Iadade
df["Idades"] = [23, 20, 22, 32, 23]
df
```

```python
# Se eu quiser criar uma coluna de e-mail e Iadades
df.loc[7,"E-mail"] = "lucas@gmail.com"
df.loc[7,"Idades"] = 23
```
